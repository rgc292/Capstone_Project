## Data Visualization

### Count of trips over months to evaluate the trend 

### FHV (For-Hire Vehicle):

```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
df_cnt2015 = pd.read_csv('2015_fhv_total_count', header=None, sep=',')
df_cnt2016 = pd.read_csv('2016_fhv_total_count', header=None, sep=',')
```

```python
df_cnt2015[0] = pd.to_datetime(df_cnt2015[0])
df_cnt2016[0] = pd.to_datetime(df_cnt2016[0])
```

```python
df_cnt2015 = df_cnt2015.set_index(0)
df_cnt2016 = df_cnt2016.set_index(0)
```

```python
df_cnt2015 = pd.pivot_table(df_cnt2015, index=df_cnt2015.index.month, columns=df_cnt2015.index.year, values=1, aggfunc='sum')
df_cnt2016 = pd.pivot_table(df_cnt2016, index=df_cnt2016.index.month, columns=df_cnt2016.index.year, values=1, aggfunc='sum')
pv = pd.concat([df_cnt2015, df_cnt2016], axis=1)
#pv.plot()
import matplotlib.patches as mpatches
plt.plot(df_cnt2015, color='#FFA500')
plt.plot(df_cnt2016, color='#f4d942')
plt.title("Count of FHV trips", fontsize=13)
plt.xlabel('Months', fontsize=13)
plt.ylabel('Number of trips', fontsize=13)
plt.axis([1, 12, .25*10000000, 1.1*10000000])
F = mpatches.Patch(color='#FFA500', label='2015', linewidth=0)
y = mpatches.Patch(color='#f4d942', label='2016', linewidth=0)
plt.legend(loc='auto', handles=[F,y])
plt.show()
```

### Uber:

```python
dfu_cnt2014 = pd.read_csv('2014_uber_total_count', header=None, sep=',')
dfu_cnt2015 = pd.read_csv('2015_uber_total_count', header=None, sep=',')
```

```python
dfu_cnt2014[0] = pd.to_datetime(dfu_cnt2014[0])
dfu_cnt2015[0] = pd.to_datetime(dfu_cnt2015[0])
```

```python
dfu_cnt2014 = dfu_cnt2014.set_index(0)
dfu_cnt2015 = dfu_cnt2015.set_index(0)
```

```python
dfu_cnt2014 = pd.pivot_table(dfu_cnt2014, index=dfu_cnt2014.index.month, columns=dfu_cnt2014.index.year, values=1, aggfunc='sum')
dfu_cnt2015 = pd.pivot_table(dfu_cnt2015, index=dfu_cnt2015.index.month, columns=dfu_cnt2015.index.year, values=1, aggfunc='sum')
pv = pd.concat([dfu_cnt2014, dfu_cnt2015], axis=1)
pv.plot()
plt.title("Count of Uber trips", fontsize=13)
plt.xlabel('Months', fontsize=13)
plt.ylabel('Number of trips', fontsize=13)
```
