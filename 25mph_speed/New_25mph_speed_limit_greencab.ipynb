## Data Evaluation

### Green Cab:

#### Null hypothesis: 
    Test 1: Ho: (_09_2014 - _09_2015) = 0
#### Alternative hypothesis:
    Test 1: HA: (_09_2014 - _09_2015) > 0
#### Level of significance:
    5% = 0.05
#### Sample size:
    346120
#### Statistical technique & corresponding test statistic to use:
    Paired t test for the mean difference in related populations.
#### Rules for rejecting or not rejecting the null hypothesis:
    Test 1: Reject the null hypothesis if the tSTAT > upper critical value, otherwise do not reject
    Test 2: Reject the null hypothesis if the p-value < alpha, otherwise do not reject

```python
import random
# Since each sample size is different, y_sample will match the size of y2
y2_sample = array(random.sample(y2,346120))
```

```python
# Test 1: Critical value approach (upper right tail)
    # alpha=0.05 , degrees_of_freedom=346120 , t(t-table)=1.645
print "Critical value= ",stats.t.ppf(1-0.05, 346120)
```

```python
stats.ttest_rel(y, y2_sample, axis=0, nan_policy='propagate')
    # t* = 41.0501855920201
    # t* > critical value , therefore , there is enough evidence to reject the null hypothesis  
```

```python
# Test 2: P value approach
    # p = 0.0 , alpha = .05
    # p < alpha , therefore , there is enough evidence to reject the null hypothesis 
```

```python
# Standard Error
np.sqrt((((y2_sample.std())**2)/346120) + (((y.std())**2)/346120))
```

```python
# Mean difference
y.mean() - y2_sample.mean()
```

### Average speed over months to evaluate the trend 

### Green Cab:

```python
df_avg2013 = pd.read_csv('2013_green_whole_Manhattan_avg', header=None, sep=',')
df_avg2014 = pd.read_csv('2014_green_whole_Manhattan_avg', header=None, sep=',')
df_avg2015 = pd.read_csv('2015_green_whole_Manhattan_avg', header=None, sep=',')
df_avg2016 = pd.read_csv('2016_green_whole_Manhattan_avg', header=None, sep=',')
```

```python
df_avg2013[0] = pd.to_datetime(df_avg2013[0])
df_avg2014[0] = pd.to_datetime(df_avg2014[0])
df_avg2015[0] = pd.to_datetime(df_avg2015[0])
df_avg2016[0] = pd.to_datetime(df_avg2016[0])
```

```python
df_avg2013 = df_avg2013.set_index(0)
df_avg2014 = df_avg2014.set_index(0)
df_avg2015 = df_avg2015.set_index(0)
df_avg2016 = df_avg2016.set_index(0)
```

```python
df_avg2013 = pd.pivot_table(df_avg2013, index=df_avg2013.index.month, columns=df_avg2013.index.year, values=1, aggfunc='sum')
df_avg2014 = pd.pivot_table(df_avg2014, index=df_avg2014.index.month, columns=df_avg2014.index.year, values=1, aggfunc='sum')
df_avg2015 = pd.pivot_table(df_avg2015, index=df_avg2015.index.month, columns=df_avg2015.index.year, values=1, aggfunc='sum')
df_avg2016 = pd.pivot_table(df_avg2016, index=df_avg2016.index.month, columns=df_avg2016.index.year, values=1, aggfunc='sum')
pv = pd.concat([df_avg2013, df_avg2014, df_avg2015, df_avg2016], axis=1)
#pv.plot()
import matplotlib.patches as mpatches
plt.plot(df_avg2013, color='b')
plt.plot(df_avg2014, color='r')
plt.plot(df_avg2015, color='#FFA500')
plt.plot(df_avg2016, color='#f4d942')
plt.title("Average speed of Green Cabs in Manhattan", fontsize=13)
plt.xlabel('Months', fontsize=13)
plt.ylabel('Average Speed', fontsize=13)
plt.axis([1, 12, 11.5, 14.6])
b = mpatches.Patch(color='b', label='2013', linewidth=0)
r = mpatches.Patch(color='r', label='2014', linewidth=0)
F = mpatches.Patch(color='#FFA500', label='2015', linewidth=0)
y = mpatches.Patch(color='#f4d942', label='2016', linewidth=0)
plt.legend(loc='auto', handles=[b,r,F,y])
plt.show()
```
