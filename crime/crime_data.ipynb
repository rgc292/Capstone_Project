# NYPD Crime Data in Manhattan

## Data Visualization

```python
import pandas as pd
df = pd.read_csv('crime_whole_Manhattan_count', sep=',', header=None)
```

```python
df_cnt2011 = df[[0,1]][:12]
df_cnt2012 = df[[0,1]][12:24]
df_cnt2013 = df[[0,1]][24:36] 
df_cnt2014 = df[[0,1]][36:48] 
df_cnt2015 = df[[0,1]][48:]
```

```python
df_cnt2011[0] = pd.to_datetime(df_cnt2011[0])
df_cnt2012[0] = pd.to_datetime(df_cnt2012[0])
df_cnt2013[0] = pd.to_datetime(df_cnt2013[0])
df_cnt2014[0] = pd.to_datetime(df_cnt2014[0])
df_cnt2015[0] = pd.to_datetime(df_cnt2015[0])
```

```python
df_cnt2011 = df_cnt2011.set_index(0)
df_cnt2012 = df_cnt2012.set_index(0)
df_cnt2013 = df_cnt2013.set_index(0)
df_cnt2014 = df_cnt2014.set_index(0)
df_cnt2015 = df_cnt2015.set_index(0)
```

```python
df_cnt2011 = pd.pivot_table(df_cnt2011, index=df_cnt2011.index.month, columns=df_cnt2011.index.year, values=1, aggfunc='sum')
df_cnt2012 = pd.pivot_table(df_cnt2012, index=df_cnt2012.index.month, columns=df_cnt2012.index.year, values=1, aggfunc='sum')
df_cnt2013 = pd.pivot_table(df_cnt2013, index=df_cnt2013.index.month, columns=df_cnt2013.index.year, values=1, aggfunc='sum')
df_cnt2014 = pd.pivot_table(df_cnt2014, index=df_cnt2014.index.month, columns=df_cnt2014.index.year, values=1, aggfunc='sum')
df_cnt2015 = pd.pivot_table(df_cnt2015, index=df_cnt2015.index.month, columns=df_cnt2015.index.year, values=1, aggfunc='sum')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
pv = pd.concat([df_cnt2011, df_cnt2012, df_cnt2013, df_cnt2014, df_cnt2015], axis=1)
pv.plot()
plt.title("Count of Crimes in Manhattan", fontsize=13)
plt.xlabel('Months', fontsize=13)
plt.ylabel('Number of crimes', fontsize=13)
```

```python
mu2011 = df_cnt2011.mean()
mu2012 = df_cnt2012.mean()
mu2013 = df_cnt2013.mean()
mu2014 = df_cnt2014.mean()
mu2015 = df_cnt2015.mean()
```

```python
mu2011.get(2011)
```

```python
import matplotlib.patches as mpatches
fig, ax = plt.subplots()
plt.axhline(y=mu2011.get(2011), xmin=0, xmax=1, hold=None, color='b')
plt.axhline(y=mu2012.get(2012), xmin=0, xmax=1, hold=None, color='g')
plt.axhline(y=mu2013.get(2013), xmin=0, xmax=1, hold=None, color='r')
plt.axhline(y=mu2014.get(2014), xmin=0, xmax=1, hold=None, color='c')
plt.axhline(y=mu2015.get(2015), xmin=0, xmax=1, hold=None, color='m')
plt.title("Avg of Crimes per Month in Manhattan", fontsize=13)
b = mpatches.Patch(color='b', label='2011', linewidth=0)
g = mpatches.Patch(color='g', label='2012', linewidth=0)
r = mpatches.Patch(color='r', label='2013', linewidth=0)
c = mpatches.Patch(color='c', label='2014', linewidth=0)
m = mpatches.Patch(color='m', label='2015', linewidth=0)
plt.legend(loc='auto', handles=[b,g,r,c,m])
plt.show()
```
