# NYPD Collision Data in Manhattan

## Data Visualization

### Count of Collisions in Manhattan

```python
import pandas as pd
df = pd.read_csv('collision_count_Manhattan', sep=',', header=None)
```

```python
df_cnt2012 = df[[0,1]][:6]
df_cnt2013 = df[[0,1]][6:18] 
df_cnt2014 = df[[0,1]][18:30] 
df_cnt2015 = df[[0,1]][30:42]
df_cnt2016 = df[[0,1]][42:]
```

```python
df_cnt2012[0] = pd.to_datetime(df_cnt2012[0])
df_cnt2013[0] = pd.to_datetime(df_cnt2013[0])
df_cnt2014[0] = pd.to_datetime(df_cnt2014[0])
df_cnt2015[0] = pd.to_datetime(df_cnt2015[0])
df_cnt2016[0] = pd.to_datetime(df_cnt2016[0])
```

```python
df_cnt2012 = df_cnt2012.set_index(0)
df_cnt2013 = df_cnt2013.set_index(0)
df_cnt2014 = df_cnt2014.set_index(0)
df_cnt2015 = df_cnt2015.set_index(0)
df_cnt2016 = df_cnt2016.set_index(0)
```

```python
df_cnt2012 = pd.pivot_table(df_cnt2012, index=df_cnt2012.index.month, columns=df_cnt2012.index.year, values=1, aggfunc='sum')
df_cnt2013 = pd.pivot_table(df_cnt2013, index=df_cnt2013.index.month, columns=df_cnt2013.index.year, values=1, aggfunc='sum')
df_cnt2014 = pd.pivot_table(df_cnt2014, index=df_cnt2014.index.month, columns=df_cnt2014.index.year, values=1, aggfunc='sum')
df_cnt2015 = pd.pivot_table(df_cnt2015, index=df_cnt2015.index.month, columns=df_cnt2015.index.year, values=1, aggfunc='sum')
df_cnt2016 = pd.pivot_table(df_cnt2016, index=df_cnt2016.index.month, columns=df_cnt2016.index.year, values=1, aggfunc='sum')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
#fig, ax = plt.subplots()
pv = pd.concat([df_cnt2012, df_cnt2013, df_cnt2014, df_cnt2015, df_cnt2016], axis=1)
#pv.plot()
plt.plot(df_cnt2012, color='c')
plt.plot(df_cnt2013, color='b')
plt.plot(df_cnt2014, color='r')
plt.plot(df_cnt2015, color='#FFA500')
plt.plot(df_cnt2016, color='y')
plt.title("Count of Collisions in Manhattan", fontsize=13)
plt.xlabel('Months', fontsize=13)
plt.ylabel('Number of collisions', fontsize=13)
plt.axis([1, 12, 50, 650])
c = mpatches.Patch(color='c', label='2012', linewidth=0)
b = mpatches.Patch(color='b', label='2013', linewidth=0)
r = mpatches.Patch(color='r', label='2014', linewidth=0)
F = mpatches.Patch(color='#FFA500', label='2015', linewidth=0)
y = mpatches.Patch(color='y', label='2016', linewidth=0)
plt.legend(loc='auto', handles=[c,b,r,F,y])
plt.show()
```

```python
mu2012 = df_cnt2012.mean()
mu2013 = df_cnt2013.mean()
mu2014 = df_cnt2014.mean()
mu2015 = df_cnt2015.mean()
mu2016 = df_cnt2016.mean()
```

```python
import matplotlib.patches as mpatches
fig, ax = plt.subplots()
plt.axhline(y=mu2012.get(2012), xmin=0, xmax=1, hold=None, color='b')
plt.axhline(y=mu2013.get(2013), xmin=0, xmax=1, hold=None, color='g')
plt.axhline(y=mu2014.get(2014), xmin=0, xmax=1, hold=None, color='r')
plt.axhline(y=mu2015.get(2015), xmin=0, xmax=1, hold=None, color='c')
plt.axhline(y=mu2016.get(2016), xmin=0, xmax=1, hold=None, color='m')
plt.title("Avg of Collisions per Month in Manhattan", fontsize=13)
g = mpatches.Patch(color='b', label='2012', linewidth=0)
r = mpatches.Patch(color='g', label='2013', linewidth=0)
c = mpatches.Patch(color='r', label='2014', linewidth=0)
m = mpatches.Patch(color='c', label='2015', linewidth=0)
b = mpatches.Patch(color='m', label='2016', linewidth=0)
plt.legend(loc='auto', handles=[b,g,r,c,m])
plt.show()
```

### Based on the city's report 116 Injuries + Fatalities occurred from 2008 to 2012. In 2013, there were 23 cases. Therefore,  it is an average of 23 cases per year from 2008 to 2014(exclusive).
### After redesigning improvements delivered on February 2014, the average declined to 13.3 ocurrences per year from 2014 to 2016. A decrease of 42%.
