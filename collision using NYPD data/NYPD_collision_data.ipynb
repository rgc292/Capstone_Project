```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
fig,plot = plt.subplots()
collision_2012 =[18,19,16,11,21,24,26,21,17,28,17,17]
collision_2013 =[15,14,17,20,22,25,19,21,21,19,28,21]
collision_2014 =[12,25,18,17,18,24,18,17,16,23,11,30]
collision_2015 =[19,18,16,15,30,23,15,20,14,22,16,22]
collision_2016 =[15,19,17,9,16,13,13,10]

plt.plot(collision_2012,label = '2012')
plt.plot(collision_2013,label = '2013')
plt.plot(collision_2014,label = '2014')  
plt.plot(collision_2015,label = '2015')
plt.plot(collision_2016,label = '2016')
plt.ylabel('number of collisions')
plt.xlabel('time')
month = ['January','March','May','July', 'September','November']
plot.set_xticklabels(month,rotation=45)
plt.title('count of collisions in Manhattan Broadway/ W135th st to W 153th st')
plt.legend(loc="lower right")
plt.show()
```

```python
collision_2012_ave = np.mean(collision_2012)
collision_2013_ave = np.mean(collision_2013)
collision_2014_ave = np.mean(collision_2014)
collision_2015_ave = np.mean(collision_2015)
collision_2016_ave = np.mean(collision_2016)
```

```python
fig,plot = plt.subplots()

plot.bar(np.arange(5),[collision_2012_ave,collision_2013_ave,collision_2014_ave,collision_2015_ave,collision_2016_ave],0.3)
year = ['2012','2013','2014','2015','2016']
plt.ylabel('number of collisions')
plt.xlabel('time')
plot.set_xticks(np.arange(5))
plot.set_xticklabels(year)
plt.title('avg of collisions per month in Manhattan Broadway/ W135th st to W 153th st')
plt.show()
```
