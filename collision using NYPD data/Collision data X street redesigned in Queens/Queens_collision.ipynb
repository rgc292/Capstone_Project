```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
fig,plot = plt.subplots()
collision_2013 =[116,90,116,110,133,115,100,128,137,145,139,139]
collision_2014 =[124,104,102,115,98,119,116,135,138,120,129,121]
collision_2015 =[109,95,95,106,101,112,121,118,122,144,98,111]
collision_2016 =[106,104,84,82,67,75,61,67]

#plt.plot(collision_2012,label = '2012')
plt.plot(collision_2013,label = '2013')
plt.plot(collision_2014,label = '2014')  
plt.plot(collision_2015,label = '2015')
plt.plot(collision_2016,label = '2016')
plt.ylabel('number of injury and fatality')
plt.xlabel('time')
month = ['January','March','May','July', 'September','November']
plot.set_xticklabels(month,rotation=45)
plt.title('Injury or Fatal Collisions per month')
plt.legend(loc="lower right")
plt.show()
```

```python
#collision_2012_ave = np.mean(collision_2012)
collision_2013_ave = np.mean(collision_2013)
collision_2014_ave = np.mean(collision_2014)
collision_2015_ave = np.mean(collision_2015)
collision_2016_ave = np.mean(collision_2016)
```

```python
fig,plot = plt.subplots()

plot.bar(np.arange(4),[collision_2013_ave,collision_2014_ave,collision_2015_ave,collision_2016_ave],0.3)
year = ['2013','2014','2015','2016']
plt.ylabel('number of injury and fatality')
plt.xlabel('time')
plot.set_xticks(np.arange(4))
plot.set_xticklabels(year)
plt.title('Average Injury or Fatal Collisions per year')
plt.show()
```
